```python
%%sql
--발전량
CREATE TABLE raw_data.generation_output (
    period VARCHAR(7),
    region VARCHAR(10),
    nuclear INT,
    bituminous_coal INT,
    sub_bituminous_coal INT,
    petroleum INT,
    LNG INT,
    reversible_pumping INT,
    fuel_cells INT,
    coal_gasification INT,
    solar INT,
    wind INT,
    hydro INT,
    ocean INT,
    bioenergy INT,
    waste INT,
    others INT,
    total INT
);
```

```python
%%sql
-- adhoc.generation_output_grouprenewables
-- 신재생 에너지 row들 합치기
CREATE TABLE adhoc.generation_output_grouprenewables AS
SELECT period,
  nuclear, bituminous_coal, sub_bituminous_coal,petroleum, lng,
  reversible_pumping + fuel_cells + coal_gasification + solar +
  wind + hydro + ocean + bioenergy + waste AS renewables,
  others, total
FROM raw_data.generation_output;
```

```python
%%sql
SELECT * FROM  adhoc.generation_output_grouprenewables
ORDER BY 1 DESC;
```

```python
%%sql
-- adhoc.generation_output_year
--연도별 그룹화 && Null 값 0으로 대치 && total은 제거
CREATE TABLE adhoc.generation_output_year AS
SELECT LEFT(period, 4) AS year,
  SUM(nuclear) AS nuclear,
  SUM(bituminous_coal) AS bituminous_coal,
  SUM(sub_bituminous_coal) AS sub_bituminous_coal,
  SUM(petroleum) AS petroleum,
  SUM(lng) AS lng,
  COALESCE(SUM(renewables), 0) AS renewables,  --Null 값 0으로 대치
  SUM(others) AS others
  -- SUM(total) AS total
FROM adhoc.generation_output_grouprenewables
GROUP BY year
ORDER BY 1 DESC;
```

```python
%%sql

SELECT * FROM adhoc.generation_output_year
ORDER BY 1 DESC;
```

```python
%%sql
-- [최종 테이블]
-- analytics.generation_output_summary
-- 2002년 ~ 2022년 데이터만 추출(시각화)
CREATE TABLE analytics.generation_output_summary AS
SELECT *
FROM adhoc.generation_output_year
WHERE year BETWEEN 2002 AND 2022
ORDER BY year DESC;
```

```python
%%sql
COMMIT;
```

```python
%%sql

SELECT * FROM analytics.generation_output_summary
ORDER BY year DESC;
```
